```python

# driver = webdriver.Chrome()
# driver.get("https://tiktak.metropolinet.co.il/#!form/222351739875465")
# driver.find_element_by_tag_name("textarea").send_keys("This is the input text which is inputed.")
# # # Function to extract input textboxes names from a webpage using Selenium

# driver = webdriver.Chrome()
# driver.get("https://tiktak.metropolinet.co.il/#!form/222351739875465")
# driver.find_element_by_tag_name("textarea").send_keys("This is the input text which is inputed.")
# # # Function to extract input textboxes names from a webpage using Selenium





def find_input_label_pairs(url):
    try:
        # Fetch the webpage content
        response = requests.get(url)
        if response.status_code != 200:
            print(f"Failed to retrieve page: {response.status_code}")
            return []

        # Parse HTML content
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find all div elements that contain both a label and an input
        divs = soup.find_all('div')
        pairs = set()  # Use a set to avoid duplicate pairs

        for div in divs:
            label = div.find('label')
            input_elem = div.find('input')

            if label and input_elem:
                label_id = label.get('id')
                input_id = input_elem.get('id')

                if label_id and input_id:
                    pairs.add((input_id, label_id))

        return list(pairs)

    except requests.RequestException as e:
        print(f"Error fetching page: {e}")
        return []

# Corrected main block
if __name__ == "__main__":
    url = "https://forms.gov.il/globaldata/getsequence/getHtmlForm.aspx?formtype=T502%40btl.gov.il"
    pairs = find_input_label_pairs(url)
    if pairs:
        print("Input and Label ID pairs found:")
        for idx, (input_id, label_id) in enumerate(pairs, 1):
            print(f"{idx}. Input ID: {input_id}, Label ID: {label_id}")
    else:
        print("No input and label ID pairs found on the page.")
```
